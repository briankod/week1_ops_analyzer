#### 1. KPI Function: Write a function calculate_efficiency(actual_output, target_output) that returns the efficiency percentage.


```python
# a function calculate_efficiency(actual_output, target_output) that returns the efficiency percentage

def calculate_efficiency(actual_output, target_output):
    percentage_output = (actual_output / target_output) * 100
    return percentage_output

efficiency_percentage  = calculate_efficiency(90, 100)
print(f"EFFICIENCY PERCENTAGE: {int(efficiency_percentage)}%")
                         
```

#### 2. Status Classifier: Write a function get_operational_status(efficiency) that returns:

- "Critical" if efficiency < 70%
- "Warning" if efficiency is between 70% and 89%
- "Normal" if efficiency >= 90%

```python
# a function get_operational_status(efficiency) that returns the status classification

def get_operational_status(efficiency):
    if efficiency >= 0.90:
        return "Normal"
    elif efficiency >= 0.70:
        return "Warning"
    else:
        return "Critical"
    
status_classification = get_operational_status(0.97)
print(F"STATUS: {status_classification}")
```

#### 3. Data Modeling: Create a dictionary representing 3 different operational sites (you choose the industry: e.g., 3 Fuel Depots, 3 Hospital Wards, or 3 Retail Stores). Each site should have keys for name , actual_output , and target_output .

```python
fuel_depots = {
    "depot_1": {
        "name": "Kenya Petroleum Refineries Limited (KPRL)",
        "actual_output": 3200,
        "target_output": 5000
    },
    "depot_2": {
        "name": "Moi International Airport",
        "actual_output": 2400,
        "target_output": 4500
    },
    "depot_3": {
        "name": "Kipevu Oil Storage Facility (KOSF)",
        "actual_output": 3500,
        "target_output": 5500
    }
}

print(list(fuel_depots.values())[1])
```

#### 4. Execution: Use a loop to iterate through your 3 sites, calculate their efficiency, determine their status, and print a formatted report for each.

```python
def formatted_report(depots):
    fuel_output = {}
    
    for site in depots.values():
        site_values = list(site.values())
        
        fuel_efficiency = calculate_efficiency(site_values[1], site_values[2])
        fuel_status = get_operational_status(fuel_efficiency)
        
        depot_name = site_values[0]
        fuel_output[depot_name] = fuel_status
        
    return fuel_output

report = formatted_report(fuel_depots)
print(f"FORMATTEED REPORT: {report}")
```
